## Import the libraries

```python
import numpy as np
import pandas as pd
```

## Load the dataset

```python
df = pd.read_csv('energydata_complete.csv')
```

```python
# Code to get information about the dataset
df
df.info()
```

```python
# Rename the columns to make it more readable
column_names = { 'date': 'date',
                'Appliances': 'Appliances',
                'lights': 'lights',
                'T1': 'Temperature_in_kitchen' , 
                'RH_1': 'Humidity_in_kitchen',
                'T2': 'Temperature_in_living_room' , 
                'RH_2': 'Humidity_in_living_room',
                'T3': 'Temperature_in_laundry_room' , 
                'RH_3': 'Humidity_in_laundry_room',
                'T4': 'Temperature_in_office_room' , 
                'RH_4': 'Humidity_in_office_room',
                'T5': 'Temperature_in_bathroom' , 
                'RH_5': 'Humidity_in_bathroom',
                'T6': 'Temperature_outside_the_building' , 
                'RH_6': 'Humidity_outside_the_building',
                'T7': 'Temperature_in_ironing_room' , 
                'RH_7': 'Humidity_in_ironing_room',
                'T8': 'Temperature_in_teenager_room_2' , 
                'RH_8': 'Humidity_in_teenager_room_2',
                'T9': 'Temperature_in_parents_room' , 
                'RH_9': 'Humidity_in_parents_room',
                'T_out': 'Temperature_outside',
                'Press_mm_hg': 'Pressure',
                'RH_out': 'Humidity_outside',
                'Windspeed': 'Windspeed',
                'Visibility': 'Visibility',
                'Tdewpoint': 'T_dew_point',
                'rv1': 'Random_variable_1',
                'rv2': 'Random_variable_2'
                }
```

```python
df = df.rename(columns = column_names)
df
```

```python
# Question 12: Fittng a linear model on the relationship between the Temperature_in_the_living_room and the Temperature_outside_the_building
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()


x = np.array(df['Temperature_in_living_room'])
y = np.array(df['Temperature_outside_the_building'])

x = x.reshape(-1, 1)

#import the train_test_split method

from sklearn.model_selection import train_test_split
x_tr, x_tst, y_tr, y_tst = train_test_split(x, y, test_size=0.3, random_state=2)

#fit the model to the training dataset
lin_reg.fit(x_tr, y_tr)

#obtain predictions
pred_values = lin_reg.predict(x_tst)


# R-Squared (R^2)
from sklearn.metrics import r2_score
r2_score = r2_score(y_tst, pred_values)
round(r2_score, 2)
```

```python
# Question 13: Normalising with the MinMaxScaler the dataset after dropping ['date', 'light'] columns

#import the MinMaxScaler

from sklearn.preprocessing import MinMaxScaler

new_df = df.drop(columns=['date', 'lights'])

scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(new_df), columns=new_df.columns)

# declaring the features (the independent variables)
features_df = normalised_df.drop(columns=['Appliances'])

# the target (the dependent variable)
appliances_target = normalised_df['Appliances']

# Now we split our dataset (70:30 train-test)
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(features_df, appliances_target, test_size=0.3, random_state=42)

# Creating our model object
linear_model = LinearRegression()

#fit the model to the training dataset
linear_model.fit(x_train, y_train)

#obtain predictions
predicted_values = linear_model.predict(x_test)

```

```python
# Question 13: Mean Absolute Error (MAE) in two decimal places

from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 2)

#From the result, the Mean Absolute Error is 0.05
```

```python
# Question 14: Residual Sum of Squares (RSS) in two decimal places

rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)

# From the result, the RSS is 45.35
```

```python
# Question 15: Root Mean Squared Error (RMSE) in three decimal places

from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)

# From the result, the RMSE is 0.088
```

```python
# Question 16: Coefficient of Determiation (R^2) in two decimal places
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)

# From the result, the Coefficient of determination is 0.15
```

```python
# Question 17: Obtain the the feature weights from the linear model

def get_weights_df(model, feat, col_name):
    # this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df


linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')

linear_model_weights

#from the dataframe vis-a-vis the options given in the question, Random_variable_1(rv1) and Windspeed lowest and highest weights respectively.
```

```python
# Question 18 :  Train a ridge regression model with an alpha value of 0.4

# import the ridge regression model

from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
ridge_predicted_values = ridge_reg.predict(x_test)
ridge_predicted_values

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, ridge_predicted_values))
round(rmse, 3)

#From the result, RMSE = 0.088 so there is no change in value when evaluated on the test set
```

```python
# Question 19: Train a lasso regression model with an alpha of 0.001 and obtain the new feature weights
# Feature Selection and Lasso Regression
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)

lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_Weight')
lasso_weights_df

# From the lasso_weights_df dataframe, only four features have non-zero feature weights
```

```python
# Question 20: The new RMSE in 3 d.p with the lasso regression
lasso_predicted_values = lasso_reg.predict(x_test)


# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_test, lasso_predicted_values))
round(rmse, 3)

# New RMSE with the lasso regression is 0.094
```
